```python
## what is the mean value of [radius, area, perimeter, concavity, compactness] of 3 measurment

mean_se = df[["radius_SE", "area_SE", "perimeter_SE", "concavity_SE", "compactness_SE"]].mean()
print(mean_se,"\n")

mean_mean = df[["radius_mean", "area_mean", "perimeter_mean", "concavity_mean", "compactness_mean"]].mean()
print(mean_mean,"\n")

mean_max = df[["radius_max", "area_max", "perimeter_max", "concavity_max", "compactness_max"]].mean()
print(mean_max,"\n")

print("sum of up is result of down... mean of (se+mean+max) of one parameter is totaly down","\n")

#radius
radius = df["radius_mean"].sum() + df["radius_SE"].sum() + df["radius_max"].sum()
print("the mean of 3 column radius is =", radius/len(df["id"]))

#area
area = df["area_mean"].sum() + df["area_SE"].sum() + df["area_max"].sum()
print("the mean of 3 column area is =", area/len(df["id"]))
# print( df["area_mean"].mean() + df["area_SE"].mean() + df["area_max"].mean(),"\n") # another way in one step

#perimeter
perimeter = df["perimeter_mean"].sum() + df["perimeter_SE"].sum() + df["perimeter_max"].sum()
print("the mean of 3 column perimeter is =", perimeter/len(df["id"]))

#concavity
concavity = df["concavity_mean"].sum() + df["concavity_SE"].sum() + df["concavity_max"].sum()
print("the mean of 3 column concavity is =", concavity/len(df["id"]))

#compactness
compactness = df["compactness_mean"].sum() + df["compactness_SE"].sum() + df["compactness_max"].sum()
print("the mean of 3 column compactness is =", compactness/len(df["id"]))
```

```python
## what is the span of [radius, area, perimeter, concavity, compactness] for mean and max 

area_mean = df["area_mean"].max() - df["area_mean"].min()
print("span of mean area is =", area_mean)

radius_mean = df["radius_mean"].max() - df["radius_mean"].min()
print("span of mean radius is =", radius_mean)

perimeter_mean = df["perimeter_mean"].max() - df["perimeter_mean"].min()
print("span of mean perimeter is =", perimeter_mean)

concavity_mean = df["concavity_mean"].max() - df["concavity_mean"].min()
print("span of mean concavity is =", concavity_mean)

compactness_mean = df["compactness_mean"].max() - df["compactness_mean"].min()
print("span of mean compactness is =", compactness_mean)

print("-------------------------------")

area_max = df["area_max"].max() - df["area_max"].min()
print("span of max area is =", area_max)

radius_max = df["radius_max"].max() - df["radius_max"].min()
print("span of max radius is =", radius_max)

perimeter_max = df["perimeter_max"].max() - df["perimeter_max"].min()
print("span of max perimeter is =", perimeter_max)

concavity_max = df["concavity_max"].max() - df["concavity_max"].min()
print("span of max concavity is =", concavity_max)

compactness_max = df["compactness_max"].max() - df["compactness_max"].min()
print("span of max compactness is =", compactness_max)
```
